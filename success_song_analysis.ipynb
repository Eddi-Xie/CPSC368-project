## Filtering

There are 40 columns and 74660 rows in top_200_df. 
Since there are more than 20 columns and 2000 rows in top_200_df, we will filter and keep columns and rows which we need for this project.


```python
# keep track_id, track_name, artist_name, release_date, week, and rank
top_200_df = top_200_df[[
    "track_id",
    "track_name",
    "artist_name",
    "release_date",
    "week",
    "rank"
]]

# standarize track_name and artist_name
top_200_df["track_name"] = top_200_df["track_name"].str.lower().str.strip()
top_200_df["artist_name"] = top_200_df["artist_name"].str.lower().str.strip()

# convert date 
top_200_df["release_date"] = pd.to_datetime(top_200_df["release_date"], errors = "coerce")
top_200_df["week"] = pd.to_datetime(top_200_df["week"], errors = "coerce")

# fliter 2020
top_200_df = top_200_df[top_200_df["week"].dt.year == 2020]

# remove duplicate
top_200_df = top_200_df.drop_duplicates(subset=["track_id"])

# remove missing track_id
top_200_df = top_200_df.dropna(subset=["track_id"])

# filtered data
top_200_df
```

## filtering

There are 21 columns and 114000 rows in spotify_track_df.
Since there are more than 20 columns and 2000 rows in the spotify_track_df, we filter and keep columns and rows which we need for this project.

```python
# keep track_id, track_name, artists, and track_genre
spotify_track_df = spotify_track_df[[
    "track_id",
    "track_name",
    "artists",
    "track_genre"
]]

# standarize track_name and artist_name
spotify_track_df["track_name"] = spotify_track_df["track_name"].str.lower().str.strip()
spotify_track_df["artists"] = spotify_track_df["artists"].str.lower().str.strip()
spotify_track_df["track_genre"] = spotify_track_df["track_genre"].str.lower().str.strip()

# remove duplicate
spotify_track_df = spotify_track_df.drop_duplicates(subset=["track_id"])

# remove missing track_id
spotify_track_df = spotify_track_df.dropna(subset=["track_id"])

# before this filtering there are still 90k rows
# filter tracks appearing in top_200_df
top_200_ids = set(top_200_df["track_id"])
spotify_track_df = spotify_track_df[spotify_track_df["track_id"].isin(top_200_ids)]

# filtered data
spotify_track_df
```

## Filtering 

There are 23 columns and 6746 rows in tiktok_df.
Since there are more than 20 columns and 2000 rows in the tiktok_df, we filter and keep columns and rows which we need for this project.

```python
# keep track_id, track_name, artist_name, release_date, and popularity
tiktok_df = tiktok_df[[
    "track_id",
    "track_name",
    "artist_name",
    "release_date",
    "popularity"
]]

# standarize track_name and artist_name
tiktok_df["track_name"] = tiktok_df["track_name"].str.lower().str.strip()
tiktok_df["artist_name"] = tiktok_df["artist_name"].str.lower().str.strip()

# convert date 
tiktok_df["release_date"] = pd.to_datetime(tiktok_df["release_date"], errors = "coerce")

# since we have a lot of data, filter release_data in 2020
# filter the song data released in 2020
tiktok_df = tiktok_df[tiktok_df["release_date"].dt.year == 2020]

# remove duplicate
tiktok_df = tiktok_df.drop_duplicates(subset=["track_id"])

# filtered data
tiktok_df
```

```python
# just check if we have enough data 

top_200_ids = set(top_200_df["track_id"])
spotify_track_ids = set(spotify_track_df["track_id"])
tiktok_ids = set(tiktok_df["track_id"])
```
